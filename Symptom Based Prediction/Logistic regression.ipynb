```python
Y=ds["COVID-19"]
X=ds.drop(["COVID-19"],axis=1)
```

```python
!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_tr,X_te,Y_tr,Y_te=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgm=LogisticRegression()
```

```python
lgm.fit(X_tr,Y_tr)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
prds=lgm.predict(X_te)
classification_report(Y_te,prds)
```

```python
confusion_matrix(Y_te,prds)
```

```python
#ACCURACY SCORE
```

```python
prd=lgm.predict(X_tr)
print("Train score:",accuracy_score(Y_tr,prd))
print("Test score:",accuracy_score(Y_te,prds))
```

```python
#HYPERPARAMETER TUNING
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

c_space = nmp.logspace(150, 280, 300)
param_grid = {'C': c_space}

logreg = LogisticRegression()
  

logreg_cv = GridSearchCV(logreg, param_grid, cv = 5)
  
logreg_cv.fit(X_te, Y_te)
 
print("Best score  LOGISTIC REGRESSION {}".format(logreg_cv.best_score_))
```
